```python
import pandas as pd
```

```python
df_cleaned_politifact_data = pd.read_csv('cleaned_politifact_data.csv')
df_cleaned_princeton_data = pd.read_csv('cleaned_princeton_data.csv')
df_afp_fact_check_clean = pd.read_csv('afp-fact-check-clean.csv')
df_ecdc_faqs_cleaned = pd.read_csv('ecdc_faqs_cleaned.csv')
df_cdc_faqs_cleaned = pd.read_csv('cdc_faqs_cleaned.csv')
df_who_myths_cleaned = pd.read_csv('who_myths_cleaned.csv')
df_CoAID_poynter_data_cleaned = pd.read_csv('CoAID_poynter_data_cleaned_13Jul_V3.csv')
df_ecdc_faqs_cleaned['claim_source'] = ""
df_cdc_faqs_cleaned['source_label'] = ""
df_who_myths_cleaned['claim_source'] = ""
```

```python
df_CoAID_poynter_data_cleaned.columns
```

```python
df_who_myths_cleaned.columns
```

```python
df1 = df_cleaned_politifact_data[['claim', 'label', 'source_label', 'source', 'date', 'claim_source','explanation', 'fact_check_url']]
df2 = df_cleaned_princeton_data[['claim', 'label', 'source_label', 'source', 'date', 'claim_source','explanation', 'fact_check_url']]
df3 = df_afp_fact_check_clean[['claim', 'label', 'source_label', 'source', 'date', 'claim_source','explanation', 'fact_check_url']]
df4 = df_ecdc_faqs_cleaned[['claim', 'label', 'source_label', 'source', 'date', 'claim_source','explanation', 'fact_check_url']]
df5 = df_cdc_faqs_cleaned[['claim', 'label', 'source_label', 'source', 'date', 'claim_source','explanation', 'fact_check_url']]
df6 = df_who_myths_cleaned[['claim', 'label', 'source_label', 'source', 'date', 'claim_source','explanation', 'fact_check_url']]
df7 = df_CoAID_poynter_data_cleaned[['claim', 'label', 'source_label', 'source', 'date', 'claim_source','explanation', 'fact_check_url']]
```

```python
frames = [df1, df2, df3,df4,df5,df6,df7]
result = pd.concat(frames)
```

```python
result['label'] = result['label'].apply(lambda x: str(x).lower())
result['source_label'] = result['source_label'].apply(lambda x: str(x).lower())
result
```

```python
result['claim'] = result['claim'].apply(lambda x: " ".join(str(x).split()))
```
