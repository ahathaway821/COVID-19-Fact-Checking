```
url = 'https://www.ecdc.europa.eu/en/covid-19/questions-answers'
req = Request(url , headers={'User-Agent': 'Mozilla/5.0'})

webpage = urlopen(req).read()
soup = BeautifulSoup(webpage, 'html.parser')
faqs = soup.find_all('h3')
```

```
d = []
for faq in faqs:
  faq_question = faq.text

  faq_concat_answers = ''
  has_answer = False
  while(1):
    faq_answer = faq.findNextSibling('p')
    if faq_answer is None:
      break
    else:
      has_answer = True
  
    faq_concat_answers = faq_concat_answers + ' ' + faq_answer.text
    faq = faq_answer

  if has_answer:
    category = 'General'
    parent_category = faq.find_previous('h2')
    if parent_category is not None and parent_category.text != 'Secondary Navigation':
      category = parent_category.text

    d.append({
        'faq_question': faq_question,
        'faq_answer': faq_concat_answers,
        'faq_category': category,
        'date_collected': date.today().strftime("%m/%d/%y")
    })

df = pd.DataFrame(d)
df
```

```
df.to_csv('ecdc_faqs.csv')
```
